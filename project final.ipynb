### Apply: General split-apply-combine concept here. Find top highest Temperature.#aggregation

```python
def top(df, n=5, column='Temperature (C)'):
    return df.sort_values(by=column)[-n:]
```

### Apply Group-Wise Linear Regression

```python
!pip install statsmodels
```

```python
import statsmodels.api as sm
```

```python
def regress(data, yvar, xvars):
    Y = data[yvar]
    X = data[xvars]
    X['intercept'] = 1.
    result = sm.OLS(Y, X).fit()
    return result.params
```

```python
by_year = data.groupby('Weather_Summary')
```

```python
by_year.apply(regress, 'Temperature (C)', ['Wind Speed (km/h)'])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=100)
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
prd=model.predict(x_test)
```

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': prd})
df
```

```python
plt.scatter(y_test,prd)
plt.xlabel('Actual Temp')
plt.ylabel('Predicted Temp')
plt.title('Actual Temp vs. Predicted Temp')
```

```python
accuracy=model.score(x_test,y_test)
print('Accuracy {:.2f}{}'.format(accuracy*100,'%'))
```
